```python
import numpy as np 
from pandas import DataFrame
import seaborn as sns
import matplotlib.pyplot as plt 
import matplotlib

matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
%matplotlib inline
```

```python
# Heatmap

Index= ['aaa', 'bbb', 'ccc', 'ddd', 'eee']
Cols = ['A', 'B', 'C', 'D']
df = DataFrame(abs(np.random.randn(5, 4)), index=Index, columns=Cols)

heatmap = sns.heatmap(df, annot=True)
plt.savefig('../text/paper/heatmap.png', format='png')
heatmap
```

```python
# Colofblind-friendly plot

x=np.arange(8)
y1=4*x+5
y2=3*x+5
y3=2*x+5
y4=x+5

colors=['bae4bc', '7bccc4', '43a2ca','0868ac']

plt.gca().set_prop_cycle(color=colors)
plt.plot(x,y1,label="4x+5")
plt.plot(x,y2,label="3x+5")
plt.plot(x,y3,label="2x+5")
plt.plot(x,y4,label="x+5")

#plt.title("Color blind-friendly palette",fontsize=15)
plt.xlabel("X",fontsize=13)
plt.ylabel("Y",fontsize=13)
plt.legend()

plt.savefig('../text/paper/color_blind_friendly_palette.png', format='png')
plt.show()
```
